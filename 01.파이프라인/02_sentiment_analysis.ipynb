## 문제02 감성분석

목표 : 주어진 문장의 긍정/부정 판단

1. I like math because it is interesting.
2. I'm against to learn coding, beause AI is better than human.

설계: 트랜스포머스 라이브러리의 sentiment-analysi 활용.(인스턴스화)

--> Distilbert 모델의 극성 판정 기능을 불러온다.

```python
from transformers import pipeline
sentiment= pipeline('sentiment-analysis')
```

```python
sentiment.model
```

```python
print(sentiment(["I like math because it is interesting."]))
```

```python
print(sentiment(["I'm against to learn coding, beause AI is better than human."]))
```
